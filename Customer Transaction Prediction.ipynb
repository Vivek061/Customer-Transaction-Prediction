```python
x = dataset.drop(['ID_code', 'target'], axis=1)
y = dataset['target']
```

```python
x.shape, y.shape
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print('x_train shape:', x_train.shape)
print('y_train shape:', y_train.shape)
print('x_test shape:', x_test.shape)
print('y_test shape:', y_test.shape)
```

```python
# logistic regression   
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
# accuracy score
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print('Accuracy:', acc)
```

```python
# classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()  
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
```

```python
dt_pred  = dt.predict(x_test)
dt_pred
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score
dt_acc = accuracy_score(y_test, dt_pred)
dt_acc
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,dt_pred))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
dt_cm = confusion_matrix(y_test, dt_pred)
print(dt_cm)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(dt_cm, annot=True, fmt='d', cmap='Blues')
plt.show()
```

```python
train_pred = dt.predict(x_train)
train_acc = accuracy_score(y_train, train_pred)
print('Training Accuracy:', train_acc)    
```

```python
print('Testing Accuracy:', dt_acc)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)
```

```python
rf_pred = rf.predict(x_test)
rf_pred
```

```python
# accuracy score
rf_acc = accuracy_score(y_test, rf_pred)
rf_acc
```

```python
# confusion matrix
cm = confusion_matrix(y_test, rf_pred)
print(cm)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues')
plt.show()
```

```python
# classification report
print(classification_report(y_test, rf_pred))
```

```python
train_pred = dt.predict(x_train)
train_acc = accuracy_score(y_train, train_pred)
print('Training Accuracy:', train_acc)
```

```python
print('Testing Accuracy:', dt_acc)
```

```python
# model comparison
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']
accuracies = [acc, dt_acc, rf_acc]
model_comparison = pd.DataFrame({'Model': models, 'Accuracy': accuracies})
model_comparison
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Model', y='Accuracy', data=model_comparison)
plt.show()
```

```python
# Feature importance
important =  rf.feature_importances_
feature_importance = pd.DataFrame({'Feature': x.columns, 'Importance': important})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
feature_importance.head(10)
```

```python
top_features = feature_importance.head(10)
plt.figure(figsize=(8,6))
sns.barplot(x='Importance', y='Feature', data=top_features)
plt.title('Top 10 Feature Importances')
plt.show()
```

```python
# Hyperparameter tuning for Random Forest
RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
}
```

```python
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_score_)
```

```python
best_rf = grid_search.best_estimator_
```

```python
best_pred = best_rf.predict(x_test)
```

```python
best_accuracy = accuracy_score(y_test,best_pred)
print(best_accuracy)
```

```python
print(classification_report(y_test,best_pred))
```

```python
from sklearn.metrics import roc_auc_score
y_prob = best_rf.predict_proba(x_test)[:,1]
y_prob
```

```python
roc_score = roc_auc_score(y_test, y_prob)

print("ROC-AUC Score:", roc_score)
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
```

```python
plt.figure(figsize=(10,8))
plt.plot(fpr, tpr)

plt.plot([0,1], [0,1], linestyle="--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.show()
```

```python
final_models = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "Tuned Random Forest"
    ],
    
    "Accuracy": [
        acc,
        dt_acc,
        rf_acc,
        best_accuracy
    ]
})

final_models
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Model', y='Accuracy', data=final_models)
plt.title('Model Comparison')
plt.show()
```
